```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Input data
fuel_qty_kg_hr = 2957
NCV_kcal_kg = 2609
heat_gen_W = 8321754.483
flue_gas_flow_kg_s = 5.71551822
Cp = 1318.319566  # J/kg-K

# Temperature assumptions
air_inlet_temp = 30                # °C
initial_T_out = 800                # Assumed outlet gas temp (°C)

# Geometry and material properties
wall_temp = 210         # °C
mean_temp = 415         # °C
emissivity = 0.357710888
refractory_area = 33.84     # m²
surface_area = 77.61        # m²
L = 1.883777864             # m
K = 0.268925279             # W/m-K
sigma = 5.67e-8             # Stefan-Boltzmann constant

# -------------------------------
# Iterative temperature update
iterations = 13
T_out = initial_T_out
T_out_list = [T_out]  # Include initial assumption (800°C)

for i in range(iterations):
    # Heat loss by conduction
    Q_cond = K * surface_area * (mean_temp - wall_temp) / L

    # Heat loss by radiation
    T_out_K = T_out + 273.15
    T_wall_K = wall_temp + 273.15
    Q_rad = emissivity * sigma * refractory_area * (T_out_K**4 - T_wall_K**4)

    # Net heat available
    Q_net = heat_gen_W - (Q_cond + Q_rad)

    # Calculate new gas outlet temperature
    T_out_new = Q_net / (flue_gas_flow_kg_s * Cp) + air_inlet_temp
    T_out_list.append(T_out_new)
    print(f"Iteration {i+1}: Outlet Gas Temp = {T_out_new:.7f} °C")

    T_out = T_out_new  # Update for next iteration

# -------------------------------
# Final result
final_T_out = round(T_out, 3)
print(f"\n Final Converged Gas Outlet Temp: {final_T_out} °C (used as Settling Chamber Inlet Temp)")

# 864.33
861.39
861.59
...
861.57524


# Plot convergence
plt.figure(figsize=(8,5))
plt.plot(range(len(T_out_list)), T_out_list, marker='o', linestyle='-', color='darkblue')
plt.title(" Outlet Gas Temperature Convergence (Starting from 800°C)")
plt.xlabel("Iteration Number")
plt.ylabel("Outlet Gas Temperature (°C)")
plt.grid(True)
plt.xticks(range(len(T_out_list)))
plt.tight_layout()
plt.show()
```

Settling Chamber Simulation

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Input data (from spreadsheet)
flue_gas_flow_kg_s = 5.71551822        # kg/s
Cp = 1310.479888                       # J/kg-K
air_inlet_temp = 861.5752447          # °C (initial gas temp entering chamber)
initial_T_out = 700                   # °C (assumed gas outlet temp to convection bank)
mean_temp = 780.7876223               # °C
wall_temp = 210                       # °C
K = 0.247167463                       # W/m-K
L = 1.693436293                       # m

# Geometric & surface properties
surface_area = 25.9                   # m²
refractory_area = 6.4                 # m²
emissivity = 0.307807109              # Dimensionless
sigma = 5.67e-8                       # W/m²-K⁴

# -------------------------------
# Iterative calculation
iterations = 13
T_out = initial_T_out
T_out_list = [T_out]

for i in range(iterations):
    # Convert temperatures to Kelvin
    T_out_K = T_out + 273.15
    T_wall_K = wall_temp + 273.15

    # Conduction loss
    Q_cond = K * surface_area * (mean_temp - wall_temp) / L

    # Radiation loss
    Q_rad = emissivity * sigma * refractory_area * (T_out_K**4 - T_wall_K**4)

    # Total heat loss
    Q_total_loss = Q_cond + Q_rad

    # Net heat carried by gas = mass flow * Cp * (T_out - air_inlet)
    Q_net = -Q_total_loss  # since losses reduce the energy
    T_out_new = Q_net / (flue_gas_flow_kg_s * Cp) + air_inlet_temp
    T_out_list.append(T_out_new)

    print(f"Iteration {i+1}: Outlet Gas Temp = {T_out_new:.4f} °C")

    T_out = T_out_new  # update for next iteration

# -------------------------------
# Final output
final_T_out = round(T_out, 4)
print(f"\n Final Mod Temp for Gas Outlet (Convection Bank Inlet): {final_T_out} °C")

# -------------------------------
# Plotting
plt.figure(figsize=(8, 5))
plt.plot(range(len(T_out_list)), T_out_list, marker='o', linestyle='-', color='teal')
plt.title(" Outlet Gas Temperature Convergence to Convection Bank")
plt.xlabel("Iteration Number")
plt.ylabel("Outlet Gas Temperature (°C)")
plt.grid(True)
plt.xticks(range(len(T_out_list)))
plt.tight_layout()
plt.show()
```

Economizer Simulation

```python
import numpy as np
import matplotlib.pyplot as plt

# Given Data from Economizer
# Heat transfer input
UA = 6461.348722            # W/K
mass_flow_gas = 5.71551822  # kg/s
mass_flow_water = 2.777777778  # kg/s
cp_gas = 1131.62145         # J/kg-K
cp_water = 4186             # J/kg-K

# Inlet temperatures
T_gas_in = 389.0484048      # °C
T_water_in = 50.0           # °C

# Initial assumptions
T_gas_out = 200.0           # °C (initial)
T_water_out = 163.2261779   # °C (initial)
iterations = 20

# Lists to store iteration history
T_gas_out_list = [T_gas_out]
T_water_out_list = [T_water_out]

# Iterative heat exchange calculation
for i in range(iterations):
    # Mean temp difference
    delta_T1 = T_gas_in - T_water_out
    delta_T2 = T_gas_out - T_water_in
    if delta_T1 == delta_T2:
        LMTD = delta_T1
    else:
        LMTD = (delta_T1 - delta_T2) / np.log(delta_T1 / delta_T2)

    # Heat transferred (Q = UA * LMTD)
    Q = UA * LMTD  # in watts

    # Update outlet temperatures
    T_gas_out = T_gas_in - Q / (mass_flow_gas * cp_gas)
    T_water_out = T_water_in + Q / (mass_flow_water * cp_water)

    # Save for plotting
    T_gas_out_list.append(T_gas_out)
    T_water_out_list.append(T_water_out)

    # Optional: print every few iterations
    print(f"Iteration {i+1}: Gas Out = {T_gas_out:.6f} °C, Water Out = {T_water_out:.6f} °C")

# Final results
print(f"\n  Final Flue Gas Outlet Temperature: {T_gas_out:.6f} °C")
print(f"Final Water Outlet Temperature: {T_water_out:.6f} °C")

# 📈 Plot 1: Flue Gas Temp Convergence
plt.figure(figsize=(10,4))
plt.subplot(1, 2, 1)
plt.plot(T_gas_out_list, marker='o', color='crimson')
plt.title("Flue Gas Outlet Temperature Convergence")
plt.xlabel("Iteration")
plt.ylabel("Temperature (°C)")
plt.grid(True)

# 📈 Plot 2: Water Temp Convergence
plt.subplot(1, 2, 2)
plt.plot(T_water_out_list, marker='s', color='teal')
plt.title("Water Outlet Temperature Convergence")
plt.xlabel("Iteration")
plt.ylabel("Temperature (°C)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

4. Steam load

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants (given)
flue_gas_mass_flow = 5.7155       # kg/s
flue_gas_inlet_temp = 389.05      # °C
flue_gas_outlet_temp = 200        # °C
cpg = 1131.62                    # J/kg·K

water_inlet_temp = 50             # °C
cpw = 4186                       # J/kg·K
overall_UA = 6461.35             # W/K
LMTD = 203.76                   # °C
economy_efficiency = 0.7          # Fraction (70%)

# Steam parameters
# Latent heat of vaporization of steam (approximate)
latent_heat_steam = 2257000       # J/kg (at 100°C)

# Simulation range for water flow rate (kg/s)
water_flow_rates = np.linspace(1, 5, 50)  # from 1 to 5 kg/s

# Arrays to store results
feed_water_temps = []
steam_loads = []
heat_absorbed = []
temp_diffs = []

for m_water in water_flow_rates:
    # Step 1: Heat available from flue gas
    delta_T_gas = flue_gas_inlet_temp - flue_gas_outlet_temp
    Q_gas = flue_gas_mass_flow * cpg * delta_T_gas  # Total heat from flue gas (W)

    # Step 2: Heat absorbed by water without efficiency loss (max possible)
    # Assuming water outlet temp = inlet temp + Q / (m * cp)
    # Let's calculate heat absorbed based on UA and LMTD (max heat transfer)
    Q_UA = overall_UA * LMTD

    # Effective heat absorbed by water (considering economy efficiency and Q_UA)
    Q_effective = min(Q_gas, Q_UA) * economy_efficiency

    # Step 3: Calculate new feed water outlet temperature
    delta_T_water = Q_effective / (m_water * cpw)
    water_outlet_temp = water_inlet_temp + delta_T_water

    # Step 4: Steam load calculation
    # Steam load = Heat absorbed / latent heat (kg/s of steam produced)
    steam_load = Q_effective / latent_heat_steam

    # Store results
    feed_water_temps.append(water_outlet_temp)
    steam_loads.append(steam_load)
    heat_absorbed.append(Q_effective)
    temp_diffs.append(delta_T_water)

# Plotting
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(water_flow_rates, feed_water_temps, 'b-', marker='o')
plt.title('Feed Water Outlet Temperature vs Water Flow Rate')
plt.xlabel('Water Flow Rate (kg/s)')
plt.ylabel('Feed Water Outlet Temp (°C)')
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(water_flow_rates, steam_loads, 'r-', marker='s')
plt.title('Steam Load vs Water Flow Rate')
plt.xlabel('Water Flow Rate (kg/s)')
plt.ylabel('Steam Load (kg/s)')
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(water_flow_rates, heat_absorbed, 'g-', marker='^')
plt.title('Heat Absorbed vs Water Flow Rate')
plt.xlabel('Water Flow Rate (kg/s)')
plt.ylabel('Heat Absorbed (W)')
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(water_flow_rates, temp_diffs, 'm-', marker='d')
plt.title('Temperature Difference in Water vs Water Flow Rate')
plt.xlabel('Water Flow Rate (kg/s)')
plt.ylabel('ΔT Water (°C)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
